```python
# compute_average()

def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average

scores=[90,50,40,30,80]
avg = compute_average(scores)

print(" Average score:",avg)
```

```python
# check_pass_fail

def check_pass_fail(score):
    if score >=50:
       return "pass"
    else:
       return "fail"

print (check_pass_fail(80))
print (check_pass_fail(30))
```

```python
# apply to five student and print results

scores = [90,50,40,30,80]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```
