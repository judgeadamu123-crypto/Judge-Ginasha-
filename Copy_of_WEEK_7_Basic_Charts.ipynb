```python
import matplotlib.pyplot as plt
```

```python
# Create student database
students = ["paul", "mark", "felix", "peter", "Alice", "Bob", "Charlie", "David", "job", "Eva"]
math_scores = [65, 80, 68, 79, 65, 85, 90, 45, 90, 75]
english_scores = [70, 85, 50, 88, 68, 78, 59, 60, 90, 80]
Computer_scores = [90, 95, 80, 85, 70, 75, 60, 65, 86, 67]
```

```python
# great a line graph adding title and axis label


plt.figure()
plt.plot(students, math_scores)
plt.title("Math Scores - Line Chart")
plt.xlabel("students")
plt.ylabel("scores")
plt.show()
```

```python
# create a bar chart
plt.figure()
plt.bar(students, math_scores)
plt.title("Math Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
# create a lie chat math
plt.figure()
plt.pie(math_scores, labels=students, autopct='%1.1f%%')
plt.title("Math Score Distribution")
plt.show()
```

```python
# great a line graph adding title and axis label


plt.figure()
plt.plot(students, english_scores)
plt.title("English Scores - Line Chart")
plt.xlabel("students")
plt.ylabel("scores")
plt.show()
```

```python
# create a bar chart
plt.figure()
plt.bar(students, english_scores)
plt.title("English  Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
# create a lie chat math
plt.figure()
plt.pie(english_scores, labels=students, autopct='%1.1f%%')
plt.title("English  Score Distribution")
plt.show()
```

```python

# great a line graph adding title and axis label


plt.figure()
plt.plot(students, Computer_scores)
plt.title("Computer Scores - Line Chart")
plt.xlabel("students")
plt.ylabel("scores")
plt.show()
```

```python
# create a bar chart
plt.figure()
plt.bar(students, Computer_scores)
plt.title("Computer   Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```
