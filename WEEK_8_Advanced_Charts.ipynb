```python
# create a histogram


plt.figure()
plt.hist(math_scores, bins=5)
plt.title("Distribution of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure()
plt.hist(english_scores, bins=5)
plt.title("Distribution of english Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure()
plt.hist(science_scores, bins=5)
plt.title("Distribution of science Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# create a scattered plot


plt.figure()
plt.scatter(math_scores, english_scores)
plt.title("Math vs English Scores")
plt.xlabel("Math Scores")
plt.ylabel("English Scores")
plt.show()
```

```python

plt.figure()
plt.scatter(science_scores, english_scores)
plt.title("science vs English Scores")
plt.xlabel("science  Scores")
plt.ylabel("English Scores")
plt.show()
```

```python

plt.figure()
plt.plot(students, math_scores)
plt.plot(students, english_scores)
plt.title("Math and English Scores Comparison")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python

plt.figure()
plt.plot(students, science_scores)
plt.plot(students, english_scores)
plt.title("science and English Scores Comparison")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python

plt.figure()

plt.subplot(1, 2, 1)
plt.hist(science_scores, bins=5)
plt.title("Science Histogram")

plt.subplot(1, 2, 2)
plt.scatter(math_scores, science_scores)
plt.title("Math vs Science")

plt.tight_layout()
plt.show()
```

```python

plt.figure()

plt.subplot(1, 2, 1)
plt.hist(english_scores, bins=5)
plt.title("english Histogram")

plt.subplot(1, 2, 2)
plt.scatter(math_scores, english_scores)
plt.title("english vs maths")

plt.tight_layout()
plt.show()
```
